# 1.9 get downsampled training sets

```python
# get files
negative = '/path/to/9_FINAL/data/machine_learning/two_class/one-hot-ecoding/train/cross_validation/negative/'
positive = '/path/to/INTENSO/9_FINAL/data/machine_learning/two_class/one-hot-ecoding/train/cross_validation/positive/'

negative_files = os.listdir(negative)
positive_files = os.listdir(positive)
```

```python
# take all positives and sample from negatives
negatives_taken = []

for x in range(5):
    # initialize empty dataframe for positives
    all_positives = pd.DataFrame()
    do_not_take_positive = 'fold_positives_' + str(x) + '.csv' 
    
    # loop through all positives
    for file in positive_files:
        if file != do_not_take_positive:
            data = pd.read_csv(positive + file, sep=";")
            all_positives = pd.concat([all_positives, data])
    
    all_negatives = pd.DataFrame()
    do_not_take_negative = 'fold_negatives_' + str(x) + '.csv'
    # loop through all negatives
    for negative_file in negative_files:
        if negative_file != do_not_take_negative:
            data = pd.read_csv(negative + negative_file, sep=";")
            all_negatives = pd.concat([all_negatives, data])
    
    # get sample of all negatives
    all_negatives = all_negatives[~all_negatives.id.isin(negatives_taken)]
    all_negatives = all_negatives.sample(len(all_positives))
    negatives_taken.extend(all_negatives.id) # do not take these instances again
    
    # stack data
    training_fold = pd.concat([all_positives, all_negatives])
    training_fold = training_fold.sample(frac=1).reset_index(drop=True)
    
    # sanity check
    print('Training {}'.format(len(training_fold)))
    
    # export
    file_out = 'training_' + str(x) + '.csv'
    training_fold.to_csv('/path/to/9_FINAL/data/machine_learning/two_class/one-hot-ecoding/train/cross_validation/training/'+file_out,sep=";")
    # print('Positives {}'.format(len(all_positives)))
```
