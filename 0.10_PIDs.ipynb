```python
import os
import pandas as pd
import numpy as np
import csv
from collections import Counter
```

# 0. Get all unique ids

```python
# get files
folder_in = '/path/to/9_FINAL/data/machine_learning/two_class/distance/merged_with_contexts_'
# os.listdir(folder_in)
```

```python
provids = []
```

```python
# get all unique ids
for i in range(1, 39):
    folder = folder_in + str(i) + '/'
    
    # get all files in the folder
    files = os.listdir(folder)
    
    for file in files:
        
        # read data and get provid
        data = pd.read_csv(folder + file, 
                          sep=";",
                          index_col=0)
        
        # debug
        # print(data.provid.tolist())
        
        # extend provids with provid
        provids.extend(data.provid.tolist())
        # print(provids)
        
    print(f'Finished with folder {str(i)}')
```

```python
# get counter from list 
counter_provids = Counter(provids)
```

```python
# provids
counter_provids
```

```python
# export counter to csv
with open('counter_provids_1.csv','w') as csvfile:
    fieldnames=['provid','count']
    writer=csv.writer(csvfile)
    writer.writerow(fieldnames)
    for key, value in counter_provids.items():
        row = str(key) + ';' + str(value)
        writer.writerow(row)
```
