# 0. Fold 0

```python
# get folders representation

folder = '/path/to/9_FINAL/data/machine_learning/two_class/distance/grouping/Run_3/validation_fold_0/count_sum_'
```

```python
all_data = pd.DataFrame()

for i in range(1,79):
    # get data for one file
    raw_data = pd.read_csv(folder+ str(i) + '.csv', sep=";")
    
    # concat data
    all_data = pd.concat([all_data, raw_data], ignore_index=True)
    
    print(len(all_data))
    print(i)
```

```python
means_0 = pd.read_csv('/path/to/9_FINAL/data/machine_learning/two_class/distance/grouping/Run_3/validation_fold_0/val_fold_0_group_mean.csv', 
             sep=";", index_col=0)
means_0
```

```python
# get ids
ids = ids[['id', 'frequency', 'pidspread', 'pldspread', 'label']]
ids_distances = pd.merge(ids, means_0,
                        how='left', left_on='id',
                        right_on='id')
ids_distances[ids_distances.p1.isna()]
```

```python
# drop observations without NA
ids_distances_WO_NA = ids_distances[~ids_distances.p1.isna()]
ids_distances_WO_NA.shape
```

```python
# drop unnecessary columns
ids_distances_WO_NA.drop(columns={'id', 'frequency',
                                 'pidspread', 'pldspread'},
                        inplace=True)
train, test = train_test_split(ids_distances_WO_NA)
y_train = train.label.to_list()
X_train = train.drop(columns={'label'})

y_test = test.label.to_list()
X_test = test.drop(columns={'label'})
```

```python
# train random Forest
model = RandomForestClassifier(random_state=42)
model.fit(X_train, y_train)
predictions = model.predict(X_train)
```

```python
predictions = model.predict(X_test)
```

```python
# print scores
print(f'Recall {recall_score(y_test, predictions)}')
print(f'Precision {precision_score(y_test, predictions)}')
print(f'F1 {f1_score(y_test, predictions)}')
```
